```python
import torch
import torch.nn as nn

class VanillaRNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(VanillaRNN, self).__init__()

        self.hidden_size = hidden_size
        self.input_size = input_size
        self.output_size = output_size

        self.input_lay = nn.Linear(input_size, hidden_size)

        self.hidden_lay = nn.Linear(hidden_size, hidden_size)

        self.output_lay = nn.Linear(hidden_size, output_size)


    def forward(self, x, h_prev):
        # x:      (batch_size, input_size)  - current input
        # h_prev: (batch_size, hidden_size) - previous hidden state

        h_t = torch.tanh(self.input_lay(x) + self.hidden_lay(h_prev))

        y_t = self.output_lay(h_t)

        return y_t, h_t

    def init_hidden(self, batch_size):
        return torch.zeros(batch_size, self.hidden_size)


# --- Test your implementation ---
if __name__ == "__main__":
    input_size  = 10
    hidden_size = 20
    output_size = 5
    batch_size  = 5
    seq_len     = 8

    model = VanillaRNN(input_size, hidden_size, output_size)

    x = torch.randn(seq_len, batch_size, input_size)
    h = model.init_hidden(batch_size)

    # TODO: loop over the sequence and collect outputs
    outputs = []
    for t in range(seq_len):
        y, h = model.forward(x[t], h)
        outputs.append(y)

    print("Done!")
    print(outputs)
```

# Implementing LSTM

```python
import torch
import torch.nn as nn

class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(LSTM, self).__init__()

        self.hidden_size = hidden_size
        self.input_size  = input_size
        self.output_size = output_size

        # TODO: define linear layers for each gate
        # Hint: each gate needs two linear layers (input → hidden, hidden → hidden)
        # You need layers for: forget, input, candidate, output gates

        # Forget gate
        self.forget_x_gate = nn.Linear(input_size, hidden_size)
        self.forget_h_gate = nn.Linear(hidden_size, hidden_size)

        # Input gate
        self.input_gate_x = nn.Linear(input_size, hidden_size)
        self.input_gate_h = nn.Linear(hidden_size, hidden_size)

        # Candidate (cell gate)
        self.candidate_gate_x = nn.Linear(input_size, hidden_size)
        self.candidate_gate_h = nn.Linear(hidden_size, hidden_size)

        # Output gate
        self.output_gate_x = nn.Linear(input_size,  hidden_size)
        self.output_gate_h = nn.Linear(hidden_size, hidden_size)

        # Final output layer (hidden → output)
        self.final_lay = nn.Linear(hidden_size, output_size)


    def forward(self, x, h_prev, c_prev):
        # x:      (batch_size, input_size)
        # h_prev: (batch_size, hidden_size)
        # c_prev: (batch_size, hidden_size)

        # TODO: compute forget gate
        # Hint: σ(linear(x) + linear(h_prev))
        f_t = torch.sigmoid(self.forget_x_gate(x) + self.forget_h_gate(h_prev))

        # TODO: compute input gate
        i_t = torch.sigmoid(self.input_gate_x(x) + self.input_gate_h(h_prev))

        # TODO: compute candidate cell state
        # Hint: use tanh instead of sigmoid
        c_candidate = torch.tanh(self.candidate_gate_x(x) + self.candidate_gate_h(h_prev))

        # TODO: compute output gate
        o_t = torch.sigmoid(self.output_gate_x(x) + self.output_gate_h(h_prev))

        # TODO: update cell state
        # Hint: c_t = f_t ⊙ c_prev + i_t ⊙ c_candidate
        c_t = f_t * c_prev + i_t * c_candidate

        # TODO: update hidden state
        # Hint: h_t = o_t ⊙ tanh(c_t)
        h_t = o_t * torch.tanh(c_t)

        # TODO: compute output
        y_t = self.final_lay(h_t)

        return y_t, h_t, c_t

    def init_hidden(self, batch_size):
        # TODO: return initial h and c states
        # Hint: now you need to return TWO zero tensors
        return torch.zeros(batch_size, self.hidden_size), torch.zeros(batch_size, self.hidden_size)


# --- Test your implementation ---
if __name__ == "__main__":
    input_size  = 10
    hidden_size = 20
    output_size = 5
    batch_size  = 5
    seq_len     = 8

    model = LSTM(input_size, hidden_size, output_size)

    x = torch.randn(seq_len, batch_size, input_size)

    # TODO: initialize h and c
    h, c = model.init_hidden(batch_size) or (None, None)

    outputs = []
    for t in range(seq_len):
        x_t = x[t]
        y_t, h, c = model.forward(x_t, h, c)
        outputs.append(y_t)

    print("Done!")
    print(outputs)
```

```python
  import torch
  import torch.nn as nn

  class GRU(nn.Module):
      def __init__(self, input_size, hidden_size, output_size):
          super(GRU, self).__init__()

          self.hidden_size = hidden_size
          self.input_size  = input_size
          self.output_size = output_size

          # TODO: define linear layers for each gate
          # GRU has only 2 gates + 1 candidate:
          #   - Reset gate  (controls how much past hidden state to use in candidate)
          #   - Update gate (controls how much to update hidden state vs keep old one)
          #   - Candidate   (new candidate hidden state)

          # Reset gate
          # self.??? = ...
          self.reset_gate_x = nn.Linear(input_size, hidden_size)
          self.reset_gate_h = nn.Linear(hidden_size, hidden_size)


          # Update gate
          # self.??? = ...
          self.update_gate_x = nn.Linear(input_size, hidden_size)
          self.update_gate_h = nn.Linear(hidden_size, hidden_size)

          # Candidate hidden state
          # self.??? = ...
          self.candidate_gate_x = nn.Linear(input_size, hidden_size)
          self.candidate_gate_h = nn.Linear(hidden_size, hidden_size)

          # Final output layer
          # self.??? = ...
          self.final_lay = nn.Linear(hidden_size, output_size)


      def forward(self, x, h_prev):
          # x:      (batch_size, input_size)
          # h_prev: (batch_size, hidden_size)
          # Note: only ONE hidden state, no cell state!

          # TODO: compute reset gate
          # Hint: same pattern as LSTM gates
          r_t = torch.sigmoid(self.reset_gate_x(x) + self.reset_gate_h(h_prev))

          # TODO: compute update gate
          z_t = torch.sigmoid(self.update_gate_x(x) + self.update_gate_h(h_prev))


          # TODO: compute candidate hidden state
          # Hint: reset gate FILTERS h_prev before passing it in
          # h_candidate = tanh(linear(x) + linear(r_t * h_prev))
          h_candidate = torch.tanh(self.candidate_gate_x(x) + self.candidate_gate_h(h_prev * r_t))

          # TODO: compute new hidden state
          # Hint: z_t blends old and new hidden state
          # h_t = (1 - z_t) * h_prev + z_t * h_candidate
          h_t = (1 - z_t) * h_prev + z_t * h_candidate

          # TODO: compute output
          y_t = self.final_lay(h_t)

          return y_t, h_t

      def init_hidden(self, batch_size):
          return torch.zeros(
              batch_size,
              self.hidden_size,
              device=next(self.parameters()).device
          )

  # --- Test your implementation ---
  if __name__ == "__main__":
      input_size  = 10
      hidden_size = 20
      output_size = 5
      batch_size  = 5
      seq_len     = 8

      model = GRU(input_size, hidden_size, output_size)

      x = torch.randn(seq_len, batch_size, input_size)
      h = model.init_hidden(batch_size)

      outputs = []
      for t in range(seq_len):
          x_t = x[t]
          y_t, h = model.forward(x_t, h)
          outputs.append(y_t)

      print("Done!")
      print(outputs)
```

```python
import torch
import torch.nn as nn

# ============================================================
# ENCODER
# Reads the entire input sequence, returns final hidden state
# ============================================================
class Encoder(nn.Module):
    def __init__(self, vocab_size, embed_size, hidden_size):
        super(Encoder, self).__init__()

        self.hidden_size = hidden_size

        # Embedding layer: converts token indices → dense vectors
        # vocab_size: number of words in source language
        # embed_size:  size of each word embedding
        self.embedding = nn.Embedding(vocab_size, embed_size)

        # TODO: define a GRU layer
        # Hint: input is embed_size, hidden is hidden_size
        # self.gru = ...
        self.gru = GRU(embed_size, hidden_size, hidden_size) # we don't need last arg???

    def forward(self, x, h):
        # x: (batch_size,) — one token index per batch item
        # h: (batch_size, hidden_size) — current hidden state

        # TODO: embed the input token
        # embedded: (batch_size, embed_size)
        embedded = self.embedding(x)

        # TODO: pass embedded + hidden through GRU
        # Hint: use your GRU forward pattern from before
        # output: (batch_size, hidden_size)
        # h_new:  (batch_size, hidden_size)
        output, h_new = self.gru.forward(embedded, h)

        return output, h_new

    def init_hidden(self, batch_size):
        return torch.zeros(
            batch_size,
            self.hidden_size
        )


# ============================================================
# DECODER
# Takes context vector, generates output sequence one token at a time
# ============================================================
class Decoder(nn.Module):
    def __init__(self, vocab_size, embed_size, hidden_size):
        super(Decoder, self).__init__()

        self.hidden_size = hidden_size

        # Embedding layer: converts token indices → dense vectors
        # vocab_size: number of words in TARGET language
        self.embedding = nn.Embedding(vocab_size, embed_size)

        # TODO: define a GRU layer
        # self.gru = ...
        self.gru = GRU(embed_size, hidden_size, hidden_size)

        # TODO: define output projection layer
        # Maps hidden_size → vocab_size (predict next word)
        # self.output_lay = ...
        self.output_lay = nn.Linear(hidden_size, vocab_size)

    def forward(self, x, h):
        # x: (batch_size,) — one token index (previous output word)
        # h: (batch_size, hidden_size) — context or previous hidden state

        # TODO: embed the input token
        embedded = self.embedding(x)

        # TODO: pass through GRU
        output, h_new = self.gru.forward(embedded, h)

        # TODO: project to vocabulary size
        # prediction: (batch_size, vocab_size)
        prediction = self.output_lay(h_new)

        return prediction, h_new


# ============================================================
# SEQ2SEQ
# Glues Encoder + Decoder together
# ============================================================
class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder):
        super(Seq2Seq, self).__init__()

        self.encoder = encoder
        self.decoder = decoder

    def forward(self, src, trg, teacher_forcing_ratio=0.5):
        # src: (src_len, batch_size) — source sequence (e.g. English)
        # trg: (trg_len, batch_size) — target sequence (e.g. French)
        # teacher_forcing_ratio: how often to use ground truth vs predicted token

        src_len, batch_size = src.shape
        trg_len, _          = trg.shape

        # TODO: initialize encoder hidden state
        h = self.encoder.init_hidden(batch_size)

        # TODO: run encoder over entire source sequence
        # loop over src_len, update h at each step
        for t in range(src_len):
            x_t = src[t]
            output, h = self.encoder.forward(x_t, h)

        # h now contains the context vector
        context_vector = h

        # TODO: initialize decoder with context vector
        # first decoder input is the <SOS> token (start of sequence)
        # trg[0] is the <SOS> token
        decoder_input = trg[0]
        decoder_hidden = h

        # TODO: run decoder, collect predictions
        outputs = []
        for t in range(1, trg_len):
            # get prediction and new hidden state
            label_t = trg[t]
            prediction, decoder_hidden = self.decoder.forward(decoder_input, decoder_hidden)
            # store prediction
            outputs.append(prediction)

            # TODO: teacher forcing
            # with probability teacher_forcing_ratio: use ground truth token
            # otherwise: use predicted token (argmax of prediction)
            use_teacher_forcing = torch.rand(1).item() < teacher_forcing_ratio
            if use_teacher_forcing:
                decoder_input = trg[t]  # ground truth
            else:
                decoder_input = prediction.argmax(dim=1)

        # stack outputs: (trg_len-1, batch_size, vocab_size)
        return torch.stack(outputs)


# ============================================================
# TEST
# ============================================================
if __name__ == "__main__":
    # Vocabulary sizes (source and target languages)
    src_vocab_size = 100
    trg_vocab_size = 120
    embed_size     = 32
    hidden_size    = 64
    batch_size     = 8
    src_len        = 6
    trg_len        = 7

    encoder = Encoder(src_vocab_size, embed_size, hidden_size)
    decoder = Decoder(trg_vocab_size, embed_size, hidden_size)
    model   = Seq2Seq(encoder, decoder)

    # Random token indices (integers, not floats)
    src = torch.randint(0, src_vocab_size, (src_len, batch_size))
    trg = torch.randint(0, trg_vocab_size, (trg_len, batch_size))

    output = model(src, trg)
    print("Output shape:", output.shape)
    # Expected: (trg_len-1, batch_size, trg_vocab_size)
    print("Done!")
```
